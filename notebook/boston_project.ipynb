```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df, x='CRIM', bins=30, kde = True, color='g')
plt.subplot(122)
sns.histplot(data=df,x='CRIM',kde=True,hue='MEDV')
plt.show()
```

```python
plt.figure(figsize = (16, 12))
sns.heatmap(df.corr(), cmap = 'Blues', annot = True, fmt = '.2%')
```

```python
corr_with_medv = df.corrwith(df['MEDV'])

plt.figure(figsize = (16, 4))
sns.heatmap([np.abs(corr_with_medv)], cmap = 'RdBu_r', annot = True, fmt='.2%')
```

```python
corr_with_medv[:-1].abs().sort_values(ascending = False)
```

```python
plt.plot(corr_with_medv[:-1].abs().sort_values(ascending = False), df.columns[:-1])
```
